```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
plt.style.use("fivethirtyeight")
%matplotlib inline
```

```python
#reading stock data from yahoo
from pandas_datareader.data import DataReader
import yfinance as yf
```

```python
#for time stamps
from datetime import datetime
```

```python
#using banking and power sector stocks for analysis
comp_list=['SHEL','NGL','HSBC','HDB']
```

```python
#Setting up end and start times for the data
end=datetime.now()
start=datetime(end.year-1,end.month,end.day)

for stock in comp_list:
    globals()[stock] =yf.download(stock,start,end)
    
```

```python
company_list=[SHEL,NGL,HSBC,HDB]
company_name=["Shell plc","NGL Energy","HSBC","HDFC"]

for company, com_name in zip(company_list,company_name):
    company["comapany_name"]=com_name
    
    
df =pd.concat(company_list, axis=0)
df.tail()
```

```python
#checking historical view of closing price


plt.figure(figsize=(15,6))
plt.subplots_adjust(top=1.25, bottom=1.2)


for i, company in enumerate(company_list,1):
    plt.subplot(2,2,i)
    company['Adj Close'].plot()
    plt.ylabel('Adj Close')
    plt.xlabel(None)
    plt.title(f"Closing Price of {comp_list[i-1]}")
    
plt.tight_layout()
```

```python
#19-07-2022


#The correlation between different stocks closing prices

closing_df = DataReader(comp_list, 'yahoo', start, end)['Adj Close']
closing_df.head()
```

```python
# Now i have all stocks closing price, so iam checking daily returns for the all stocks

daily_rets = closing_df.pct_change()
daily_rets.head()
```

```python
# i will use a jointplot to compare the daily returns of SHELL and NGL

sns.jointplot(x='SHEL', y='NGL', data=daily_rets, kind='scatter')
```

```python
# i will use a jointplot to compare the daily returns of HSBC and HDB

sns.jointplot(x='HSBC', y='HDB', data=daily_rets, kind='scatter')
```

```python
# Now iam using a pairplot to analyze all companies data

sns.pairplot(daily_rets, kind='reg')
```

```python
#now iam visualizing correlation plot for daiy retuns 

sns.heatmap(daily_rets.corr(), annot=True , cmap='hot')
```

```python
#now iam visualizing correlation plot for closing price

sns.heatmap(closing_df.corr(), annot=True , cmap='hot')
```
